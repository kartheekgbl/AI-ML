```python
#importing modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import Data
data=pd.read_csv('D:\\github-d\\AI-ML\\Datasets\\50_Startups.csv')
data
```

```python
# Split as X and y
X=data.drop('Profit',axis=1)
y=data['Profit']
```

```python
# Handel categorical data
states=pd.get_dummies(X,drop_first=True)
states
```

```python
#Drop State column in X
X=states
X
```

```python
# Concat the Categorical numaric data to X
# X=pd.concat([X,states],axis=1)
# X
```

```python
#Standerdization
Xt=((X-X.mean())/X.std())
#y=((y-y.mean())/y.std())
Xt=np.array(Xt)
Xt
```

```python
# For Prediction
def predict_y(w,x,b):
    ylst=[]
    for i in range(len(x)):
        ylst.append(w@x[i]+b)
    return np.array(ylst)

# For Loss
def loss(y,y_pred):
    n=len(y)
    s=0
    for i in range(n):
        s+=(y[i]-y_pred[i])**2
    return (1/n)*s 


# For change in weight(gradient dw) 

def dldw(x,y,y_pred):
    n=len(y)
    s=0
    for i in range(n):
        s+=-x[i]*(y[i]-y_pred[i])
    return (2/n)*s


# For change in weight(gradient db)  

def dldb(y,y_pred):
    n=len(y)
    s=0
    for i in range(n):
        s+=-(y[i]-y_pred[i])
    return (2/n)*s



#Model started

def Gradient(x,y):
    
    w=np.random.randn(x.shape[1])
    lr=0.001
    b=0
    iters=2000
    loss_lst=[]
    for i in range(iters):
        
        # predict y
        y_pred=predict_y(w,x,b)
        print(y_pred)
        # Loss
        loss_lst.append(loss(y,y_pred))
        
        # dw
        w=w-lr*dldw(x,y,y_pred)
        
        # db
        b=b-lr*dldb(y,y_pred)    
    
    # Plotting graph between iterations and loss 
    plt.plot(np.arange(1,iters),loss_lst[1:])
    plt.xlabel("No.of Iterations")
    plt.ylabel("Loss")       
    
    return w,b,loss_lst


w,b,loss_lst=Gradient(Xt,y)
        
```

```python
def predict(input):
    y_lst=[]
    for i in range(len(input)):
        y_lst.append(w@input[i]+b)
    return y_lst
y_pred=predict(Xt)

print(y_pred)
        
```
