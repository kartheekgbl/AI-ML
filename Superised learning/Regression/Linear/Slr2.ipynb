```python
# Implimentation of linear regression without sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("D:\\github-d\\AI-ML\\Datasets\\advertising.csv")

X=df.iloc[:,0]
y=df.iloc[:,1]
X,y
df
```

```python
plt.scatter(X,y)
plt.show()
```

```python

# Standardization for y
X_mean = X.mean()
std_X = X.std()
X = (X - X_mean) / std_X

# Standardization for y
y_mean= y.mean()
std_y = y.std()
yt = (y - y_mean) / std_y

X,y

```

```python

# Gradient descent
X,y
m=0
c=0
lr=0.1
iter=1000

for i in range(iter):
    y_pred=m*X+c
    dm=(-2/len(X))* sum(X*(y-y_pred))
    dc=(-2/len(X))*sum(y-y_pred)
    m-=lr*dm
    c-=lr*dc

print(m,c)
```
