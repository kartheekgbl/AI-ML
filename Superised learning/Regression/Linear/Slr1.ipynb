```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel("D:\\github-d\\AI-ML\\Datasets\\Employee2.xlsx")

X=np.array(df['Experience'])
y=np.array(df['Salary'])

X,y
```

```python
plt.scatter(X,y)
plt.show()
```

```python
X_mean=X.mean()
y_mean=y.mean()

num=0
den=0
for i in range(len(X)):
    num+=(X[i]-X_mean)*(y[i]-y_mean)
    den+=(X[i]-X_mean)**2
m=num/den
c=y_mean - m * X_mean
ypred=m*X+c
ypred

kk=pd.DataFrame()
kk['actual']=y
kk['predict']=ypred
kk
```

```python
plt.scatter(X,y)
plt.plot(X,ypred,c="red")
plt.show()
```

```python
num1=0
den1=0
for i in range (len(X)):
    num1+=(y[i]-ypred[i])**2
    den1+=(y[i]-y_mean)**2
rsq=1-(num1/den1)
rsq
```

```python
from sklearn.metrics import r2_score
r2_score(y,ypred)
```
